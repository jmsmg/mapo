```python
import pandas as pd
import folium
```

```python
df = pd.read_csv('서울시 지정·인증업소 현황.csv', encoding='euc-kr') #데이터프레임 불러오는 부분
df.isnull().sum() #Null값, 열 확인
```

```python
df_1 = df[['업소 명', '업태 명', '지도 X좌표', '지도 Y좌표','도로명상세주소']] #필요부분 추출
df_1
```

```python
m = folium.Map(
    location=[37.56619800217754, 126.90164250057964],
    zoom_start=[14]
)

for i in range(df.shape[0]):
    folium.Marker(
        [df.iloc[i]['지도 Y좌표'], df.iloc[i]['지도 X좌표']],
        popup = f"<strong>{df.iloc[i]['업소 명']}</strong>"
    ).add_to(m)

tooltip1 = 'Click'
folium.Marker(
    [37.581888972371935, 126.88949066574179],
    popup = '집',
    icon = folium.Icon(color = '#ff00000', icon='home')
).add_to(m)
m
```
